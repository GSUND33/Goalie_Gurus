```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
from splinter import Browser
from bs4 import BeautifulSoup
import urllib.request
import time
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prior_seasons = pd.read_csv("nhl_data_before_this_season.csv")

```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
seasons_tables_list = []
for i in range(2020,2021):
    try:
        new_list = []
        season_for_table = []
        year_season = f'{i-1}-{i}'
        url= f'https://www.hockey-reference.com/leagues/NHL_{i}_skaters.html#stats::goals'
        browser.visit(url)
        if i == 2020:
            time.sleep(30)
        else:
            time.sleep(10)
        html = browser.html
        zek = pd.read_html(html,header=1)
        new_df = zek [len(zek)-1]
        new_df = new_df[new_df.Rk != ('Rk')]
        new_df = new_df.reset_index()
        for j in range(len(new_df)):
            season_for_table.append(year_season)
        for k in range(len(new_df)):
            if int(new_df.iloc[k,0]+1) == int(new_df.iloc[k,1]):
                new_list.append(True)
            else:
                new_list.append(False)
        new_df ['Season'] = season_for_table
        new_df ["Keep"] = new_list
        new_df = new_df[new_df.Keep != False]
        seasons_tables_list.append(new_df)
        time.sleep(1)
    except:
        print(f'Year{i} attempted')
current_season = seasons_tables_list[0]
current_season.to_csv("this_season.csv")
```

```python
all_seasons = pd.concat([current_season,prior_seasons],sort=False)
all_seasons['seasonid'] = all_seasons.groupby(['Player']).cumcount(ascending=False) + 1 

all_seasons
```
